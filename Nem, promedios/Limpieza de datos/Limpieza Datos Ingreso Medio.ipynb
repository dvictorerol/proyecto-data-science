```python
import pandas as pd
import numpy as np
```

```python
ingreso_promedio = pd.read_excel("ingreso-medio-mensual-por-región-2010---2021.xlsx")
```

```python
ingreso_promedio
```

```python
ingreso = ingreso_promedio.iloc[0:207]
```

```python
ingreso.drop("Unnamed: 0",axis=1, inplace = True)
```

```python
ingreso = ingreso[2:]
```

```python
ingreso.rename(columns={"Unnamed: 1":"agno","Unnamed: 2":"region","Unnamed: 3":"ingre_medio","Unnamed: 4":"inter_inf","Unnamed: 5":"inter_sup","Unnamed: 6":"calidad"}, inplace = True)
```

```python
ingreso.drop(2, axis=0, inplace = True)
```

```python
ingreso[["ingre_medio", "inter_inf", "inter_sup"]] = ingreso[["ingre_medio", "inter_inf", "inter_sup"]].astype(np.float64)
ingreso["region"] = ingreso["region"].replace("Región de Tarapacá", 1)
ingreso["region"] = ingreso["region"].replace("Región de Antofagasta", 2)
ingreso["region"] = ingreso["region"].replace("Región de Atacama", 3)
ingreso["region"] = ingreso["region"].replace("Región de Coquimbo", 4)
ingreso["region"] = ingreso["region"].replace("Región de Valparaíso", 5)
ingreso["region"] = ingreso["region"].replace("Región de O’Higgins", 6)
ingreso["region"] = ingreso["region"].replace("Región del Maule", 7)
ingreso["region"] = ingreso["region"].replace("Región del Biobío", 8)
ingreso["region"] = ingreso["region"].replace("Región de La Araucanía", 9)
ingreso["region"] = ingreso["region"].replace("Región de Los Lagos", 10)
ingreso["region"] = ingreso["region"].replace("Región de Aysén", 11)
ingreso["region"] = ingreso["region"].replace("Región de Magallanes", 12)
ingreso["region"] = ingreso["region"].replace("Región Metropolitana", 13)
ingreso["region"] = ingreso["region"].replace("Región de Los Ríos", 14)
ingreso["region"] = ingreso["region"].replace("Región de Arica y Parinacota", 15)
ingreso["region"] = ingreso["region"].replace("Región del Ñuble", 16)
ingreso["region"] = ingreso["region"].replace("Nacional", 0)

ingreso["region"] = ingreso["region"].astype("category")
```

```python
ingreso
```

```python
ingreso.to_csv('Ingresos_medios_2010-2021.csv', index=False)
```
